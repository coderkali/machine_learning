```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("titanic_sample_1000.csv")
dataset.head(3)
```

```python
sns.histplot(x="Age",data=dataset)
plt.show()
```

Positive Skew Chart because its a greater than 0

```python
dataset["Age"].skew()
```

```python
dataset["Age"].mean()
```

```python
dataset["Age"].median() 
```

```python
dataset["Age"].mode()[0]
```

Nagative Skew Chart because its a less than 0

```python
data = np.random.normal(0,100,100)
data
```

```python
df = pd.DataFrame({"x":data})
df["x"].skew()
```

#Mean Median #Mode

```python
df["x"].mean(), df["x"].median(), df["x"].mode()
```

```python
sns.histplot(x="x",data=df)
plt.show()
```
