Finding out the Range 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("titanic_sample_1000.csv")
dataset.head()
```

```python
min_r = dataset["Age"].min()
```

```python
max_r = dataset["Age"].max()
```

```python
dataset[dataset["Age"] > 100]
```

```python
range = max_r - min_r
range
```

Mean Aboslute Deviation

```python
sec_a = np.array([75,65,73,68,72,76])
sec_b = np.array([90,47,43,96,93,51])
```

```python
mean_sec_A = np.mean(sec_a)
```

```python
mean_sec_B = np.mean(sec_b)
```

```python
no = np.array([1,2,3,4,5,6])
```

```python
plt.figure(figsize=(10,3))
plt.scatter(sec_a,no, label="Sec A")
plt.scatter(sec_b,no, label="Sec B")
plt.plot([70,70,70,70,70,70],no,c="blue", label="Mean")
plt.legend()
plt.show()
```

Matematical way of iodentyifying the MAD

```python
mad_a = np.sum(np.abs(sec_a-mean_sec_A))/len(sec_a)
mad_a
```

```python
mad_b = np.sum(np.abs(sec_b-mean_sec_B))/len(sec_b)
mad_b
```
