```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("titanic_sample_1000.csv")
dataset.head()
```

```python
dataset.head(3)
```

```python
dataset["Age"].mean()
```

```python
mn = np.mean(dataset["Age"])
```

```python
sns.histplot(x="Age",data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)])
plt.axvline(mn, color='red', linestyle='--', label=f'mean = {mn:.2f}')
plt.show()
```

```python
dataset.isnull().sum()
```

```python
dataset["Age"] = dataset["Age"].fillna(dataset["Age"].mean())
```

```python
dataset["Age"].median()
```

```python
fareMean = dataset["Fare"].mean()
fareMedian = dataset["Fare"].median()
fareMean, fareMedian
```

```python
sns.histplot(x="Fare",data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([fareMean for i in range(0,300)],[i for i in range(0,300)], c = "red")
plt.plot([fareMedian for i in range(0,300)],[i for i in range(0,300)], c = "blue")
plt.show()
```

```python
dataset["Age"].mode()
```

```python
fareMode = dataset["Fare"].mode()[0]
```

```python
dataset["Fare"].value_counts()
```

```python
sns.histplot(x="Fare",data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([fareMean for i in range(0,300)],[i for i in range(0,300)], c = "red", label = "mean")
plt.plot([fareMedian for i in range(0,300)],[i for i in range(0,300)], c = "blue", label = "median")
plt.plot([fareMode for i in range(0,300)],[i for i in range(0,300)], c = "green",label = "mode")
plt.legend()
plt.show()
```
