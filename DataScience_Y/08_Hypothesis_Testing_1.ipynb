## Hypothesis Testing (Z-Test) - Two Sample

H0 : mu_new = mu_old
Ha : mu_new > mu_old (right-tailed test)

Z_test = [ (xbar_new - xbar_old) - (mu_new - mu_old) ] / (sigma / sqrt(n))

condition: n >= 30 (large sample, use Z-test not t-test)

```python
import scipy.stats as st
import numpy as np
```

Example 2 : e-commerce company (XYZ), check if new website design increased avg purchase amount vs old design.

30 customers - old design, 30 customers - new design. Population std = 2.5

```python
Old_design_data = np.array([45.2, 42.8, 38.9, 43.5, 41.0, 44.6, 40.5, 42.7, 39.8, 41.4,
                             44.3, 39.7, 42.1, 40.6, 43.0, 42.2, 41.5, 39.6, 44.0, 43.1,
                             38.7, 42.5, 40.9, 41.8, 43.4, 39.9, 42.0, 40.2, 44.1, 41.2])

New_design_data = np.array([48.5, 49.1, 50.2, 47.8, 48.7, 49.9, 48.0, 50.5, 49.8, 49.6,
                             48.2, 48.9, 49.7, 50.3, 49.4, 50.1, 48.6, 48.3, 49.0, 50.0,
                             48.4, 49.3, 49.6, 48.8, 50.4, 49.2, 48.1, 49.5, 50.0, 48.9])
```

```python
pop_std = 2.5
n_sp= len(New_design_data)
apl = 0.05
```

```python
mean_new = np.mean(New_design_data)
mean_old = np.mean(Old_design_data)
```

```python
z_cal = (mean_new-mean_old)/(pop_std/np.sqrt(n_sp))
z_cal
```

```python
z_table = st.norm.ppf(0.95)
z_table
```

```python
if z_cal > z_table:
    print("ha is right")
else:
    print("h0 is right")
```
