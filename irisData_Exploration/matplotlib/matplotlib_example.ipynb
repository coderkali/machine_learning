```python
import matplotlib.pyplot as plt

# 10 students heights in cm
heights = [142, 155, 148, 162, 155, 158, 142, 165, 148, 155]

fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(heights, bins=5, color='steelblue', edgecolor='white')
ax.set_title('Student Heights')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Number of Students')
plt.show()
```

```python
# All 10 values: 142, 155, 148, 162, 155, 158, 142, 165, 148, 155

# Min = 142
# Max = 165
# Range = 165 - 142 = 23cm
# Bin width = 23 / 5 = 4.6cm

# Bin 1 → 142.0 to 146.6 → 142, 142          → 2 students
# Bin 2 → 146.6 to 151.2 → 148, 148          → 2 students
# Bin 3 → 151.2 to 155.8 → 155, 155, 155     → 3 students
# Bin 4 → 155.8 to 160.4 → 158               → 1 student
# Bin 5 → 160.4 to 165.0 → 162, 165          → 2 students
```
