```python
# Plain Python — no NumPy
revenue = [1200, 1500, 900, 1800, 2100, 2400, 1950]

# Total
total = 0
for r in revenue:
    total = total + r
print("Total:", total)

# Average
average = total / len(revenue)
print("Average:", average)

# Best day
best = revenue[0]
for r in revenue:
    if r > best:
        best = r
print("Best day:", best)

# Worst day
worst = revenue[0]
for r in revenue:
    if r < worst:
        worst = r
print("Worst day:", worst)
```

```python
# Plain Python — same tasks
branch_1 = [1200, 1500, 900, 1800, 2100, 2400, 1950]
branch_2 = [800,  900, 750, 1200, 1800, 2000, 1600]

# On how many days was branch 1 better?
days_won = 0
for i in range(len(branch_1)):
    if branch_1[i] > branch_2[i]:
        days_won = days_won + 1
print("Days branch 1 won:", days_won)

# Branch 1 revenue ONLY on days it beat branch 2
winning_days = []
for i in range(len(branch_1)):
    if branch_1[i] > branch_2[i]:
        winning_days.append(branch_1[i])
print("Branch 1 revenue on winning days:", winning_days)
```

```python
revenue = np.array([1200, 1500, 900, 1800, 2100, 2400, 1950])

# Pick specific days — Monday, Wednesday, Saturday
# Plain Python way
plain = [1200, 1500, 900, 1800, 2100, 2400, 1950]
specific_plain = [plain[0], plain[2], plain[5]]
print("Plain Python specific days:", specific_plain)

# NumPy way — pass a list of positions directly
specific_numpy = revenue[[0, 2, 5]]
print("NumPy specific days:", specific_numpy)

# Only days where revenue was above 1500
print("Good days only:", revenue[revenue > 1500])

# Only days where revenue was between 1000 and 2000
print("Medium days:", revenue[(revenue > 1000) & (revenue < 2000)])
```

```python
revenue = np.array([1200, 1500, 900, 1800, 2100, 2400, 1950])

# & means AND — both conditions must be true
print("Between 1000 and 2000:", revenue[(revenue > 1000) & (revenue < 2000)])

# | means OR — at least one condition must be true
print("Below 1000 OR above 2000:", revenue[(revenue < 1000) | (revenue > 2000)])

# ~ means NOT — flips True to False and False to True
print("above 2000:", revenue[(revenue > 2000)])

print("not above 2000:", revenue[~(revenue > 2000)])
```
