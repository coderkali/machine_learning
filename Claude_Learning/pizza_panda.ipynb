"Which pizza type made the most money last week?"

```python
#First — plain Python way so you feel the pain:

# Plain Python — total quantity per pizza type
pizzas = ['Margherita', 'Pepperoni', 'BBQ Chicken']

for pizza in pizzas:
    total = 0
    for i in range(len(data['pizza'])):
        if data['pizza'][i] == pizza:
            total += data['quantity'][i]
    print(pizza, ":", total)
```

```python
print(df.groupby("pizza")["quantity"].sum())
```

```python
print("Total Quantity :: ",df.groupby('pizza')['quantity'].sum())
print("\naverage quantity ::" ,df.groupby('pizza')['quantity'].mean()) 
print("\nhighest price :: ",df.groupby('pizza')['price'].max())     
print("\nhow many orders ::",df.groupby('pizza')['quantity'].count())
```

```python
# Which area — Downtown or Suburbs — generates more total revenue?

df['revenue'] = df['price'] * df['quantity']
print(df)
```

```python
# Which area generates more total revenue?
print("\nGenerates more total revenue ::" ,df.groupby('area')['revenue'].sum()) 
```

```python
# What is the total revenue AND average quantity per pizza type?
print(df.groupby('pizza').agg({
    'revenue': 'sum',
    'quantity': 'mean'
}))
```
